```python
# Install required libraries
#!pip install gtts

from gtts import gTTS
from IPython.display import Audio, display
import tempfile

def speak(text):
    """Convert text to speech and play it in Google Colab."""
    tts = gTTS(text)
    with tempfile.NamedTemporaryFile(delete=False, suffix=".mp3") as fp:
        tts.save(fp.name)
        return Audio(fp.name, autoplay=True)

# Sample questions
questions = [
    {"question": "What is 2 + 2?", "options": ["3", "4", "5"], "answer": "4"},
    {"question": "What is the capital of India?", "options": ["Delhi", "Mumbai", "Kolkata"], "answer": "Delhi"},
    {"question": "Solve: 12 * 8", "options": ["96", "88", "108"], "answer": "96"},
]

def register_user():
    """Register user and check if they are a Person with Disability (PWD)."""
    name = input("Enter your name: ")
    pwd_status = input("Are you a Person with Disability (yes/no)? ").strip().lower()
    is_pwd = pwd_status == 'yes'
    print(f"Welcome {name}!")
    if is_pwd:
        print("Accessibility features are enabled.")
        display(speak(f"Welcome {name}, accessibility features are enabled."))
    else:
        print("Accessibility features are disabled.")
    return name, is_pwd

def ask_question(q, use_audio=False):
    """Ask a question and collect the user's answer."""
    print("\nQuestion:", q["question"])
    for i, opt in enumerate(q["options"], 1):
        print(f"{i}. {opt}")

    if use_audio:
        display(speak(q["question"]))

    answer = input("Your answer: ").strip()
    return answer, q["answer"]

def start_mcq_test(audio_enabled):
    """Conduct an MCQ test."""
    score = 0
    for q in questions:
        user_ans, correct_ans = ask_question(q, audio_enabled)
        if user_ans.lower() == correct_ans.lower():
            score += 1
    return score, len(questions)

def give_feedback(score, total):
    """Provide feedback based on the user's performance."""
    percent = (score / total) * 100
    print(f"\n----- Assessment Report -----\nScore: {score} out of {total}")

    if percent >= 80:
        msg = "Excellent! You have strong skills."
    elif percent >= 50:
        msg = "Good! Some improvement needed."
    else:
        msg = "Needs Improvement. Keep learning!"

    print(msg)
    display(speak(msg))

def start_viva():
    """Placeholder for viva mode."""
    print("\nüó£Ô∏è Viva Mode Coming Soon!")
    display(speak("Viva mode is under development. Please check back later."))

def start_practical():
    """Placeholder for practical mode."""
    print("\nüõ†Ô∏è Practical Mode Coming Soon!")
    display(speak("Practical mode is under development. Please check back later."))

def main():
    """Main function to run the assessment tool."""
    name, is_pwd = register_user()

    print("\nSelect Mode:")
    print("1. MCQ Test\n2. Viva\n3. Practical")

    choice = input("Enter your choice (1/2/3): ").strip()

    if choice == '1':
        score, total = start_mcq_test(is_pwd)
        give_feedback(score, total)
    elif choice == '2':
        start_viva()
    elif choice == '3':
        start_practical()
    else:
        print("Invalid choice.")
        if is_pwd:
            display(speak("Invalid choice. Please try again."))

# Run the main function
main()
```
